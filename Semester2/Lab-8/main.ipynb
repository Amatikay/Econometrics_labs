Для предложенного временного ряда (Рождаемость):
- провести сглаживание методом скользящего среднего
- выделить линейный тренд
- найти индексы сезонности
- оценить тренд-сезонную
- построить прогноз ожидаемого среднего значения следующего элемента ряда
- провести декомпозицию ряда, используя встроенную процедуру в R по примеру товарооборота.


```r
ts <- read.csv(file = "../Data/ts_dates.csv", sep ="\t", header = FALSE)
```

```r
head(ts)
```

```r
plot(ts$V3, type='l')
```

```r
smoothed <- numeric()
for (i in 1:length(ts$V3+ 1)){ # Прсотое сколбзящее среднее с периодом 3
    smoothed[i] <- mean(ts$V3[i:(i + 3)]) 
}
```

```r
plot(ts$V3, type='l')

time_index <- 1:length(ts$V3)

model <- lm(ts$V3 ~ time_index)
summary(model)
trend <- predict(model)

lines(trend, col = "red")
lines(smoothed, col = "blue")
```

```r
library(car)   
library(forecast)

turnover=ts
t_o=ts(turnover,frequency=12,start=c(2008,1))
# t_o
plot.ts(t_o)
```

```r
logto=log(t_o) # Переход в логарифимческую шкалу
plot.ts(logto)
```

```r
decomp=decompose(logto) # Разложение на тренд, сезонность и остатки
# decomp
plot(decomp)
```

```r
seasonIndex <- decomp$seasonal # Индексы сезонности
plot(seasonIndex)
```

```r
# Оценка тренд-сезонной компоненты
trend_seasonal <- decomp$trend + seasonIndex

# Визуализация оценки тренд-сезонной компоненты
plot(trend_seasonal)
```

```r
res=decomp$random 
res=as.vector(na.omit(res)) # Обработка NA
hist(res)
```

```r
shapiro.test(res) # Нормальность остатков

dwt(as.vector(res)) # Discrete Wavelet Transform
```

```r
toforecasts = HoltWinters(t_o, gamma=1) # Прогнозирование

plot(toforecasts)
```

```r
toforecasts2 = forecast(toforecasts, h=6)
plot(toforecasts2)
# toforecasts2
```

```r
arima_e=arima(res,order=c(0,0,0),include.mean=FALSE,seasonal=list(order=c(1,0,0),period=3))
arima_e
summary(arima_e)
```

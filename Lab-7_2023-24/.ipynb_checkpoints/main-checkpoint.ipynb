Создаю `dummy variable` для *district*.

Теперь я могу использовать линейную регрессию для анализа влияния таких перменных, как:
- Количество комнат
- Район
- Площадь
<br>

На целевую перменную: `Стоимость аренды`

```r
unique_districts <- sort(unique(data$district))
for (district in unique_districts) {
 data[[paste0("district_", district)]] <- ifelse(data$district == district, 1, 0)}

data[ , -7] # Таблица без исходной перменной

pairs(data[ , -c(1,4,6,7)],pch=18) 
```

 Не самое частое а смаое дорогое среднее

```r
sum(data$district_1)
sum(data$district_2)
sum(data$district_3)
sum(data$district_4)
```

Район 3 встречается чаще всех. Базовое значение района - 3.

```r
NDummyReg <- lm(data$rent ~ (data$rooms + data$s + data$district_1 + data$district_2 + data$district_4) )
summary(NDummyReg)
```

Исключаю из модели количество комнат

```r
NDummyReg <- lm(data$rent ~ (data$s + data$district_1 + data$district_2 + data$district_4))
summary(NDummyReg)
```

<h1>Вывод</h1>

Значащим параметром является `площадь`. 
<br>
Относительно `3` района имеют значимые отличия по стоимости районы `1` и `2`.

Если предположить, что этот датасет соответствует реальным данным недвижимости в Томске, зная, что по стоимости разделяют Кировский | Cоветский ( дорогие ) 
; 
Ленинский | Октябрьский (дешевые) 
можно утверждать, что этот датасет не для Томской недвижимости 😄.

Район `1` относительно `3` снижает стоимость недвижимости на **864** ± **386**<br>
Район `2` относительно `3` увеличивает стоимость недвижимости на **1640** ± **416**<br>
Район `4` относительно `3` аналогичны.<br>
